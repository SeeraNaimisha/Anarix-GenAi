````python
def get_sql_from_question(question):
    prompt = f"""Convert this question into an SQL query for a SQLite database. Only output the SQL query.
    The available tables are:
    - eligibility (columns: eligibility_datetime_utc, item_id, eligibility, message)
    - ad_sales (columns: date, item_id, ad_sales, impressions, ad_spend, clicks, units_sold)
    - total_sales (columns: date, item_id, total_sales, total_units_ordered)

    Question: {question}
    """
    response = model.generate_content(prompt)
    sql = re.sub(r"```sql\n|```", "", response.text.strip())
    return sql
````

````python
def query_sql(query):
    conn = sqlite3.connect("ecommerce_data.db")
    cursor = conn.cursor()
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        columns = [description[0] for description in cursor.description]
        return [dict(zip(columns, row)) for row in result]
    except Exception as e:
        return {"error": str(e)}
    finally:
        conn.close()
````

````python
def ask_agent(question):
    sql = get_sql_from_question(question)
    print("Generated SQL:\n", sql)
    result = query_sql(sql)
    return result


ask_agent("What is my total sales?")
````

````python
import sqlite3
import pandas as pd

conn = sqlite3.connect("ecommerce_data.db")

roas_df = pd.read_sql("""
SELECT
    ROUND(SUM(ad_sales) * 1.0 / SUM(ad_spend), 2) AS RoAS
FROM ad_sales
WHERE ad_spend > 0;
""", conn)


cpc_df = pd.read_sql("""
SELECT
    item_id,
    ROUND(SUM(ad_spend) * 1.0 / SUM(clicks), 2) AS CPC
FROM ad_sales
WHERE clicks > 0
GROUP BY item_id
ORDER BY CPC DESC
LIMIT 1;
""", conn)

print("RoAS:\n", roas_df)
print("\nHighest CPC Product:\n", cpc_df)
conn.close()
````
